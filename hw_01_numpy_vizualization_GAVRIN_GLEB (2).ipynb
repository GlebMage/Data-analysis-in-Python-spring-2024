```python
def prod_and_sq_sum(A):
    diag1 = np.diag(A)
    q1 = diag1 ** 2
    q1P = np.prod(q1)
    s1P = np.sum(q1)
    return q1P, s1P
```

```python
np.random.seed(61)
list_time_mean = []
for floor in range(0,31):
    dist_from_zero2 = floor/30
    place = np.random.rand(10000)
    dist_place = abs(place - dist_from_zero2)
    time_St = dist_place * 30
    mean_time = np.mean(time_St)
    list_time_mean.append(mean_time)
arr_time_mean = np.array(list_time_mean)    
print(arr_time_mean)    
    
    
```

```python
np.random.seed(34)
groups_total = []
for exp in range(1000):
    pars = np.random.uniform(0, 1, 100)
    lead_index = 0
    groups = 1
    for par in range(1, 100):
        if pars[lead_index] >= pars[par]:
            groups += 1
            lead_index = par
    groups_total.append(groups)
arr_groups_total = np.array(groups_total)    
print(np.mean(arr_groups_total))
    
```
